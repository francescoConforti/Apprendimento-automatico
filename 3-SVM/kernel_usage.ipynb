```python
import math

def gaussian_kernel(xx, yy):
    mat = np.zeros((len(xx), len(yy)))
    for i in range(len(xx)):
        for j in range(len(yy)):
            mat[i,j] = gaussian(xx[i], yy[j])
    return mat

def gaussian(x, y):
    if len(x) != len(y):
        return None
    sigma=1
    sub=[]
    for i in range(len(x)):
        sub.append(x[i]-y[i])
    norm=0
    for s in sub:
        norm += s**2
    norm = math.sqrt(norm)
    return math.exp(- norm / (2*sigma**2))

def cubic_kernel(xx, yy):
    mat = np.zeros((len(xx), len(yy)))
    for i in range(len(xx)):
        for j in range(len(yy)):
            mat[i,j] = cubic(xx[i], yy[j])
    return mat

def cubic(x, y):
    if len(x) != len(y):
        return None
    dot = x @ y
    dot += 1
    return dot**3

def phi_kernel(xx, yy):
    mat = np.zeros((len(xx), len(yy)))
    for i in range(len(xx)):
        for j in range(len(yy)):
            mat[i,j] = phi(xx[i], yy[j])
    return mat

def phi(x, y):
    xvec = [x[0]**2, math.sqrt(2)*x[0]*x[1], x[1]**2]
    yvec = [y[0]**2, math.sqrt(2)*y[0]*y[1], y[1]**2]
    s = 0
    for i in range(len(xvec)):
        s += xvec[i] * yvec[i]
    return s

def gaussian_kernelV2(xx, yy):
    mat = np.zeros((len(xx), len(yy)))
    for i in range(len(xx)):
        for j in range(len(yy)):
            mat[i,j] = gaussianV2(xx[i], yy[j])
    return mat

def gaussianV2(x, y): # different value for sigma
    if len(x) != len(y):
        return None
    sigma=0.25
    sub=[]
    for i in range(len(x)):
        sub.append(x[i]-y[i])
    norm=0
    for s in sub:
        norm += s**2
    norm = math.sqrt(norm)
    return math.exp(- norm / (2*sigma**2))
```

```python
# https://jakevdp.github.io/PythonDataScienceHandbook/05.07-support-vector-machines.html

def plot_svc_decision_function(model, ax=None, plot_support=True):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    
    # plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0],
                   model.support_vectors_[:, 1],
                   s=300, linewidth=1, facecolors='none');
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

```python
clf = SVC(kernel=gaussian_kernel, C = 2.5).fit(X, target_1)
plt.scatter(X[:,0].tolist(), X[:,1].tolist(), c=labels_1, marker='o')
# CAN'T ACCESS SUPPORT VECTORS WITH CUSTOM KERNEL (SKLEARN BUG)
#plt.scatter(clf.support_vectors_[:,0], clf.support_vectors_[:,1])
plot_svc_decision_function(clf, plot_support=False);
```

```python
clf = SVC(kernel=cubic_kernel, C = 100).fit(X, target_2)
plt.scatter(X[:,0].tolist(), X[:,1].tolist(), c=labels_2, marker='o')
plot_svc_decision_function(clf, plot_support=False);
```

```python
clf = SVC(kernel=gaussian_kernelV2).fit(X, target_3)
plt.scatter(X[:,0].tolist(), X[:,1].tolist(), c=labels_3, marker='o')
plot_svc_decision_function(clf, plot_support=False);
```
