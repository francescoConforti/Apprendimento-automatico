# Exercise 1
Implement the AdaBoost ensemble algorithm by completing the following code:

````python
import copy 

class SVC_:
    def __init__(self, kernel="rbf", degree="3"):
        self.svc = svm.SVC(kernel=kernel, degree=degree, gamma='auto')

    def fit(self, X,y,sample_weight=None):
        if sample_weight is not None:
            sample_weight = sample_weight * len(X)

        self.svc.fit(X,y,sample_weight=sample_weight)
        return self

    def predict(self, X):
        return self.svc.predict(X)

def error(ground, prediction):
    return 0.5 - np.dot(ground.T, prediction)/(2*len(ground))

class AdaBoost:
    def __init__(self, weakModel, T):
        self.A = weakModel
        self.T = T
        self.m = [None]*T
        for i in range(T):
            self.m[i] = copy.deepcopy(weakModel)
            
        self.a = np.full(T, 1.)
        
        self.M = None
        
    def fit(self, X, y, verbose=False):
        w_err = [1.]
        ens_error = []
        
        n = len(X)
        y = np.array(y)        
        w = np.full(n, 1./float(n))
        
        for t in range(self.T):
            if not verbose:
                print(f"Iter {t+1}/{self.T}\tw_err={w_err[-1]:.2f}, a[t-1]={self.a[max(t-1, 0)]:.2f}\r", end='', flush=True)
            
            self.m[t].fit(X, y, sample_weight=w)
            preds = self.m[t].predict(X)
            
            #from scikit source https://github.com/scikit-learn/scikit-learn/blob/7389dbac82d362f296dc2746f10e43ffa1615660/sklearn/ensemble/weight_boosting.py#L549
            #incorrect = preds != y
            ## Error fraction
            #e = np.mean(
            #    np.average(incorrect, weights=sample_weight, axis=0)
            #)

            e = np.dot(w, ((y != preds).astype('float')))
            w_err.append(e)
            
            self.a[t] = 0.5 * np.log((1.-e)/e)
       
            w = w * np.exp(-self.a[t]*y*preds)       
            w = w / np.linalg.norm(w, 1)
            
            if t % 10 == 0:
                err = error(y, np.sign(
                    np.dot(self.a[:t+1], np.array(list(map(lambda m: m.predict(X), self.m[:t+1]))))
                ))
                ens_error.append(err)
            
            if verbose:
                print(f"Iter {t+1}/{self.T}")
                print("e: ", e)
                print("(1-e)/e:", (1.-e)/e)
                print("a: ", self.a[t])
                print("w: ", w)
                print("err: {err}")
            
        print("\nDone!")
        
        self.M = lambda X: np.sign(
            np.dot(self.a, np.array(list(map(lambda m: m.predict(X), self.m))))
        )
        
        plt.figure(figsize=(10, 8))
        plt.plot(np.arange(0, self.T, 1), w_err[1:], label='w_err')
        plt.plot(np.arange(0, self.T, 10), ens_error, label='ens_error (log)')
        plt.plot(np.arange(0, self.T, 1), self.a, label='alpha')
        plt.legend()
        plt.show()
        
        return self
        
    def predict(self, X):
        if self.M is None:
            print('Error: you must call fit() before predict()')
        return self.M(X)
````

In the implementation you are free to assume:
- that the problem is a binary classification problem with labels in $\{-1, +1\}$.
- that the weakModel can fit a weighted sample set by means of the call `weakModel.fit(X,y,sample_weight=w)` where `w` is a vector of length $|y|$.

**Note 1**:  
since the labels are bound to be in ${+1, -1}$, the classification error can be easily computed as:
$$
   error(y,y') = \frac{1}{2} - \frac{y^T \times y'}{2N},
$$
where $N$ is the total number of examples. The formula can be derived noticing that $y^T \times y'$ calculates the number $N_c$ of examples correctly classified  minus the number $N_{\bar c}$ of examples incorrectly classified. We have then $y^T \times y' = N_c - N_{\bar c}$ and by noticing that $N = N_c + N_{\bar c}$:
$$
   N - y^T \times y' = 2 N_{\bar c} \Rightarrow \frac{N - y^T \times y'}{2 N} = \frac{N_{\bar c}}{N} = error(y,y')
$$

# Exercise 2

Write a weak learner to be used with the AdaBoost algorithm you just wrote. The weak learner that you will implement shall work as follows:

- creates a random linear model $y(x) = \mathbf{w} \cdot \mathbf{x} + t$ by generating the needed weight vector $\mathbf{w}$ and $t$ at random; each weight shall be sampled from U(-1,1);
- it evaluates the weighted loss $\epsilon_t$ on the given dataset and flip the linear model if $\epsilon_t > 0.5$
- at prediction time it predicts +1 if $\mathbf{x} \cdot \mathbf{w} > 0$ it predicts -1 otherwise.

````python
class RandomLinearModel:
    def __init__(self):
        self.w = None
        self.t = None
        
    def loss(self, y, y_, w):
        if w is None:
            w = np.full_like(y, 1.)
            
        return np.dot(w, (np.array(y) != np.array(y_).astype('float')))
        
    def fit(self,X,y,sample_weight=None):
        self.w = np.random.uniform(low=-1., high=-1, size=X.shape[1]+1)
        p = self.predict(X)
        l = self.loss(y, p, sample_weight)

        if l > 0.5:
            self.w = -self.w
            
        return self
        
    def predict(self,X):
        return np.sign(
            np.dot(self.w, np.insert(X, 0, 1, axis=1).T)
        )
````
